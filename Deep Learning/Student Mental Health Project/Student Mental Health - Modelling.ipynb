# Libraries

```python
import numpy as np
import tensorflow as tf
```

# Loading the datasets

```python
npz = np.load('Mental_Health_train.npz')

# training datasets
train_inputs = npz['inputs']
train_targets = npz['targets']

# validation datasets
npx = np.load('Mental_Health_val.npz')
validation_inputs = npz['inputs']
validation_targets = npz['targets']

# testing datasets
npx = np.load('Mental_Health_test.npz')
test_inputs = npz['inputs']
test_targets = npz['targets']
```

# Model

```python
# Defining Layer size
INPUT_LAYER_SIZE = 7
HIDDEN_LAYER_SIZE = 50
OUTPUT_LAYER_SIZE = 3

# Creaing a neural network model with 4 layer depth
model = tf.keras.Sequential([
    tf.keras.layers.Dense(HIDDEN_LAYER_SIZE, activation='relu'),
    tf.keras.layers.Dense(HIDDEN_LAYER_SIZE, activation='relu'),
    tf.keras.layers.Dense(HIDDEN_LAYER_SIZE, activation='relu'),
    tf.keras.layers.Dense(OUTPUT_LAYER_SIZE, activation='softmax')
])

# Compiling the model on accuracy matrix
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Defining the batch size and number of epochs (no. of iterations)
BATCH_SIZE = 100
NUM_EPOCHS = 50

# Early stopping for avoiding overfitting 
EARLY_STOPPING = tf.keras.callbacks.EarlyStopping(patience=2)

# Fitting the model on training set and validating on validation set
model.fit(
    train_inputs,
    train_targets,
    batch_size=BATCH_SIZE,
    callbacks=[EARLY_STOPPING],
    validation_data=(validation_inputs, validation_targets),
    epochs=NUM_EPOCHS,
    verbose=2
)

# Prediction on test set and testing the test accuracy
test_loss, test_accuracy = model.evaluate(test_inputs, test_targets)
print(f"\nTest Loss: {test_loss:.2f}, Test Accuracy: {test_accuracy*100:.2f}%")

# Saving the trained model
model.save('Mental_State.keras')
```
