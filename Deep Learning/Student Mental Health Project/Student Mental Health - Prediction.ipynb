# Libraries

```python
import numpy as np
from sklearn import preprocessing
import tensorflow as tf
import pandas as pd
from IPython.display import clear_output
```

# Loading the model

```python
model = tf.keras.models.load_model('Mental_State.keras')
```

# Getting Student Data

```python
# Platform mapping
platform_map = {
    'facebook': 0,
    'tikTok': 1,
    'youTube': 2,
    'whatsapp': 3,
    'snapchat': 4,
    'instagram': 5,
    'twitter': 6,
}

def get_platform_code(platform_name):
    """
    Returns the mapped integer value for a given platform name.
    If the platform is not found, returns None.
    """
    return platform_map.get(platform_name.lower(), None) 

# Gender Encoding
def gender_encoder(value):
    if value.lower() == 'female':
        return 0
    elif value.lower() == 'male':
        return 1
    else:
        return 0


['gender', 'daily_screen_time_min', 'social_media_time_min',
       'negative_interactions_count', 'positive_interactions_count',
       'anxiety_level', 'stress_level'],

def user_input():
    print('Enter all the values below: ')
    gender = gender_encoder(input('Gender of the user (Male, Female, Other): '))
    daily_screen_time_min = int(input('Total daily device screen time in minutes: '))
    social_media_time_min = int(input('Total time spent on social media in minutes per day: '))
    negative_interactions_count = float(input('Number of negative or harmful interactions experienced online: '))
    positive_interactions_count = float(input('Number of positive or supportive interactions experienced online: '))
    anxiety_level = float(input('Self-reported anxiety score (scale 1-4): '))
    stress_level = float(input('Self-reported stress score (scale 5-9): '))
    
    return [[
        gender,
        daily_screen_time_min,
        social_media_time_min,
        negative_interactions_count,
        positive_interactions_count,
        anxiety_level,
        stress_level,
    ]]

# Getiing user data
user_data = user_input()

# Making prediction
prediction = model.predict(np.array(user_data))

# Displaying result
mental_state_map = {
    0: 'Healthy',
    1: 'Stressed',
    2: 'At Risk'
}
predicted_class = np.argmax(prediction, axis=1)
clear_output(wait=True)
print (f"The Student's Mental State is: {mental_state_map.get(predicted_class[0])}")
```
